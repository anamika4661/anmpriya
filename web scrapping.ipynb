```python
from bs4 import BeautifulSoup
import requests
final_data = [] 

for i in range(1,4):
    url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/', 'html.parser')
    #------------ request --------------------------------#
    response = requests.get(url)
    #print(response)
    data = response.text
    data = BeautifulSoup(data, "xml")
    data.table
    data = data.find_all('tr')
    realdata = data[1:-3]
    # ----------------------------filtering requiered html tag-------------------#
    output = []   
    for i in (realdata):
        output.append(i.get_text())
    #print(output) -------- get data from tag
    for i in output:
        x = i.split('\n')[1:-1]
        final_data.append(x)
final_data
```

```python
import requests
from bs4 import BeautifulSoup
final_data = [] 

# You can also get data for specific countries by changing the URL
# For example
# for US: https://www.worldometers.info/coronavirus/country/us/
# for India: https://www.worldometers.info/coronavirus/country/india/

url = "https://www.worldometers.info/coronavirus/"
req = requests.get(url)
bsObj = BeautifulSoup(req.text, "html.parser")
bsObj.table
bsObj = bsObj.find_all('tr')
output = []   
for i in (bsObj):
    output.append(i.get_text())
for i in output:
    x = i.split('\n')[2:-11]
    y = f'{x}'
    final_data.append(y)
final_data = final_data[0:230]
final_data



```

```python
import csv
with open('worldometer1.csv', 'w', newline='') as wm:
    writer = csv.writer(wm)
    for i in range(0,len(final_data)):
        z = final_data[i]
        writer.writerow(z)
```
