```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
response = requests.get(url)
# ----------------------------------REQUEST----------------------------------
data = response.text
data =BeautifulSoup(data, "xml")
data = data.table
data1 = data.find_all('tr')
# -------------------------------------Filtering required html tag------------------
data = data1[1:]
data2=data[0:-3]
output = []
for i in data2:
  output.append(i.get_text())

# -----------------------------------Get data from tag---------------------------

final_data = []
for i in output:
  x = i.split('\n')[1:-1]
  final_data.append(x)
final_data
# --------------------------------------Put the data in list inside list----------
```

```python
from bs4 import BeautifulSoup
import requests
final_data = []
n = int(input("Enter page no = "))
for i in range(1,3):
  url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/', 'html.parser')
  response = requests.get(url)
  # ----------------------------------REQUEST----------------------------------
  data = response.text
  data =BeautifulSoup(data, "xml")
  data = data.table
  data1 = data.find_all('tr')
  # -------------------------------------Filtering required html tag------------------
  data = data1[1:]
  data2=data[0:-3]
  output = []
  for i in data2:
    output.append(i.get_text())

  # -----------------------------------Get data from tag---------------------------

  
  for i in output:
    x = i.split('\n')[1:-1]
    final_data.append(x)
final_data
# --------------------------------------Put the data in list inside list----------
```
